```python
import os
from xml.etree import ElementTree
from os import listdir
from os.path import isfile, join
# xml to dict package
from PIL import Image
```

```python
myPathHelper = os.getcwd() + '/images'                             
folderList = listdir(myPathHelper)
```

```python
file_name_xml = 'templateXML.xml'
full_file_xml = os.path.abspath(os.path.join('assets', file_name_xml))

myTree = ElementTree.parse(full_file_xml)
myRoot = myTree.getroot()

xmlEnding = '.xml'
```

```python
!pip install deepface
from deepface import DeepFace
import cv2
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
for folder in folderList:
   
    mypath = myPathHelper + '/' + folder                        
    myFiles = [f for f in listdir(mypath) if f.endswith('.jpg')]
    
    print(folder)
    
    for file in range(len(myFiles)):
        
        file_name = myFiles[file]
        full_file = mypath + '/' + file_name
        
        print(full_file)
        obj = DeepFace.analyze(img_path = full_file, enforce_detection=False)
        
        myRoot[0].text = folder
        myRoot[1].text = file_name
        myRoot[2].text = full_file
        myRoot[6][0].text = folder
        
        myRoot[6][4][0].text = str(obj.get('region').get('x'))
        myRoot[6][4][1].text = str(obj.get('region').get('y'))
        myRoot[6][4][2].text = str((obj.get('region').get('x') + obj.get('region').get('w')))
        myRoot[6][4][3].text = str(obj.get('region').get('y') + obj.get('region').get('w'))
        
        im = Image.open(full_file)
        for xSize in range(2):
            myRoot[4][xSize].text = str(im.size[xSize])
            #print(str(im.size[xSize]))
        file_name_cut = file_name.split('.jpg')
        newName = file_name_cut[0] + xmlEnding
        #print(os.path.join(mypath, newName))
        myTree.write(os.path.join(mypath, newName))
```
